#Cleaning data

```python
# check how many null values in our data set
df.isna().sum()
```

```python
# Removing the missing value rows in the dataset
df.drop("App Name", inplace= True, axis=1)
df.drop("Rating", inplace= True, axis=1)
df.drop("Rating Count", inplace= True, axis=1)
df.drop("Installs", inplace= True, axis=1)
df.drop("Minimum Installs", inplace= True, axis=1)
df.drop("Currency", inplace= True, axis=1)
df.drop("Size", inplace= True, axis=1)
df.drop("Minimum Android", inplace= True, axis=1)
df.drop("Developer Id", inplace= True, axis=1)
df.drop("Developer Website", inplace= True, axis=1)
df.drop("Developer Email", inplace= True, axis=1)
df.drop("Released", inplace= True, axis=1)
df.drop("Privacy Policy", inplace= True, axis=1)
```

```python
df.drop("Last Updated", inplace= True, axis=1)
df.drop("Content Rating", inplace= True, axis=1)
df.drop("Scraped Time", inplace= True, axis=1)
```

```python
df.fillna(df.mean(),inplace=True)
```

```python
df.isnull().values.any()# checking the missing value it Removing or not 
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop("App Id", inplace= True, axis=1)
```

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

#split Data

```python
# detect input and output
X = df.drop(['Editors Choice_False','Editors Choice_True'],axis=1)
```

```python
y = df['Editors Choice_False']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=30)

X_train
```

```python
# feature scaling

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_test_scaled
```

```python
X_train_scaled
```

#Modling 

```python
# logistic regression
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train_scaled,y_train)
```

```python
y_pred = lr.predict(X_test_scaled)

y_pred
```

```python
y_test
```

Evaluation

```python
from sklearn.metrics import classification_report

print(classification_report(y_pred,y_test))
```

_____

```python
# applay Random Forest

from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier()
rf.fit(X_train_scaled,y_train)
```

```python
y_pred_rf= rf.predict(X_test_scaled)

print(classification_report(y_pred_rf,y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train_scaled,y_train)

y_pred_knn = knn.predict(X_test_scaled)

from sklearn.metrics import classification_report

print(classification_report(y_pred_knn,y_test))
```
